```python
import pandas as pd
```

```python
df = pd.read_csv('../data/melb_data.csv')
```

```python
df1 = df[['Suburb', 'Rooms', 'Price','Date','Bathroom', 'Landsize']]
```

```python
df1 = df1[~df1.Landsize.isnull() & ~df1.Price.isnull() & ~df1.Bathroom.isnull()]
df1 = df1[df1.Landsize > 100]
df1 = df1[df1.Landsize < 201]
df1.head(100).to_csv('../data/hs.csv', columns=['Rooms', 'Price','Bathroom', 'Landsize'], header=False, index=False)
```
